```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pts=np.array([[2,14],[3,12],[1,11],[3,15],[5,14],[4,12],[5,15],[2,11]])
```

```python
plt.scatter(pts[:,0],pts[:,1])
```

```python
grid=[[m,c] for m in range(15) for c in range(15)]
```

```python
print(grid)
```

```python
def get_loss(pts,m,c):
    loss=0;
    for pt in pts:
        diff=pt[1]-(m*pt[0]+c)
        loss += (diff)*(diff)
        return loss
```

```python
for (m,c) in grid:
    print(get_loss(pts,m,c),m,c)
```
